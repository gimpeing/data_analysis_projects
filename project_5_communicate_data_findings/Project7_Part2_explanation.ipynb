## Investigation Overview

In this investigation, I wanted to look at the Citi Bike ridership trend varies before and during Covid-19 Pandemic, that is trend in 2019 (pre-Covid-19) and 2020. The main focus was on key influence of the change in user type: Subscriber (Annual Member) and Customer (1-day or 3-day user). Variables with trends varies based on the user type were investigated. They are trip duration, frequency of ridershp by day of week and time of day, and the popular route (defined by start station and end station). 

## Dataset Overview

The data consisted of 708780 Citi Bike rides from 2019 to Oct 2020. They are information about individual rides made, including variables like the trip duration, start station name, end station name, user type, gender, birth year, bike id and others. About 0.9% of data were removed from the analysis due to abnormal value, for example user age of more than or equal to 80 years old. 


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

# for zipfile, combining files
import zipfile
import os
import glob

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset and combine the individual csv file into a pandas dataframe
folder_loc = './data/'

# use `glob()` to return a list of all files with `.csv` extention, save in `all_files`
folder_dir = folder_loc + 'csv_file/'
all_files = glob.glob(folder_dir + '*.csv')
print(f'Total csv files: {len(all_files)}')

# concatenate all files into a pandas dataframe
df = pd.concat([pd.read_csv(f) for f in all_files])
```

```python
# data cleaning

# change to timestamp format
df['starttime'] = pd.to_datetime(df['starttime'])
df['stoptime'] = pd.to_datetime(df['stoptime'])

# change id related and gender fromat from numeric to string
df['start station id'] = df['start station id'].astype(str)
df['end station id'] = df['end station id'].astype(str)
df['bikeid'] = df['bikeid'].astype(str)
df['gender'] = df['gender'].astype(str)

# decode the gender to unknown, male and female
df['gender'] = df['gender'].replace({'0':'unknown', '1':'male', '2':'female' })

# remove user with age >=80 years old
# add a column for user's age
df['age'] = df['starttime'].dt.year - df['birth year']

# drop row with trip duration exceeding 2 hours (equilvalent to 120 min)
df['tripduration'] = df['tripduration']/60 # convert to min
df = df.query('tripduration<=120')
df.reset_index(drop=True, inplace=True)
```

## Distribution of Ridership

There are two user types in Citi Bike data system, **Subscriber** (Annual Member) and **Customer** (1-day pass or 3-day pass user). Total number of users plummeted in year 2020 but there a surge in Customer user in 2020.

```python
# add year and month columns
df['year'] = df['starttime'].dt.year
df['month'] = df['starttime'].dt.month

# total riders by year
ridership_count = df['year'].value_counts()
ridership_order = df['year'].value_counts().index
```

```python
# color settings
colors_yr = ['#ff9999','#66b3ff']
colors_usertype = ['#ff9999', '#99ff99']
```

```python
# Total Ridership by year
g = sns.FacetGrid(data = df, col = 'year', hue = 'usertype',
                 height = 4, palette=colors_usertype)
g.map(sns.countplot, 'year', alpha = 0.6)

g.set(xticklabels = [])  # remove xticks
g.set_xlabels("")
g.add_legend();

# add main title
plt.subplots_adjust(top = 0.8)
g.fig.suptitle("Total Ridership by Year and User Type", fontsize = 16);
```

## Distribution of Ridership by month

- Apr 16.  : statewide PAUSE order extended through May 15
- June 8.  : Phase 1 reopening
- June 22. : Phase 2 reopen

```python
# Now, break down to monthly ridership, by each year
month_list = ['Jan','Feb', 'Mar', 'Apr', 'May', 'Jun',
             'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

g = sns.FacetGrid(data = df, row = 'year', hue = 'usertype',
                 height = 3, aspect = 2, palette=colors_usertype)
g.map(sns.countplot, 'month', alpha = 0.6)
g.add_legend()

# set xticks
for ax in g.axes.ravel():
    ax.set_xticklabels(month_list, rotation = 90)
plt.subplots_adjust(hspace = 0.2)
g.set_xlabels("")

# add main title
plt.subplots_adjust(top = 0.9)
g.fig.suptitle("Total Ridership by Month and User Type", fontsize = 16);
```

## Distribution of Trip Duration

Trip duration take on a very large range of values, but 75 percentile of the trip duration is < 14.3 min. However, there are outliers with extreme long trip duration, max up to approximately 37 days, which is not realistics. In addition, users need to pay extra if they exceed the time limit (30 mins for Customer, 45 mins for Subscriber). With this in mind, user typically won't exceed the total single riding time by 2 hours. Thus, the dataset remove the entries with trip duration exceeding 2 hours.   
Bar plot showed the overall trip duration increases in 2020, regardless of user type.

```python
# Average trip duration by year and user type
plt.figure(figsize = [8,5])

sns.barplot(data = df, x = 'year', y = 'tripduration', hue='usertype',
           palette=colors_usertype, ci = 'sd');
plt.ylabel("Avg. Trip Duration (min)")
plt.title("Average Trip Duration by Year and User Type", fontsize = 16);
```

## Distribution of Trip Duration by Month

Histogram (faceting by month) was plotted on a log scale due to long tail in normal distribution. In year 2020, distribution started to shift to right (longer trip duration) from April. Relatively significant shift observed in the month of **May-Aug**.

```python
# due to long tail in normal distribution, the trip duration is put on log scale instead
log_binsize = 0.05
bins = 10** np.arange(0.002, np.log10(df['tripduration'].max())+log_binsize, 
                      log_binsize)
```

```python
# Trip duration by year, month
ticks = [1, 5, 10, 50, 100, 200]
labels = [1, 5, 10, 50, 100, 200]

g = sns.FacetGrid(data = df, col = 'month', hue = 'year',
                 col_wrap = 3, height = 5, palette=colors_yr)
g.map(plt.hist, 'tripduration', bins = bins, alpha = 0.6).set(xscale = 'log')

# set x, y labels and legend
g.set_xlabels("Trip duration (min)")
g.set_ylabels("Count")
g.add_legend()

# set xticks
for ax in g.axes:
    ax.set_xticks(ticks)
    ax.set_xticklabels(labels)
plt.subplots_adjust(hspace = 0.3)
#g.set(xticks = ticks, xticklabels = labels)


# add main title
plt.subplots_adjust(top = 0.9)
g.fig.suptitle("Trip Duration Distribution by Month", fontsize = 20);
```

## Trip Frequency by Time

First, the plot showed the trip frequency by day of week. 
Day of week trend shifted from year 2019 to 2020, with higher frequency from weekdays to weekend instead.

```python
# Facetint by month of the year, countplot for trips frequency by day of week
g = sns.FacetGrid(data = df, col = 'month', hue = 'year',
                 col_wrap = 3, height = 5, palette=colors_yr)
g.map(sns.countplot, 'day_of_week', alpha = 0.6)

g.add_legend()
g.set_xlabels("");

# set xticks
for ax in g.axes.ravel():
    ax.set_xticklabels(day_list, rotation = 90)
plt.subplots_adjust(hspace = 0.2)

# add main title
plt.subplots_adjust(top = 0.9)
g.fig.suptitle("Number of Ridership by Day of Week", fontsize = 20);
```

## Popular Route

This is feature engineered variable, where the route refering to **start station** to  **end station** location. There are **two** types of route, route with different start and end station, and **round trip** where start and end station are the same.   
Top 10 popular route type shifted, from non round trip (10/10)to **round trip** (7/10) in year 2020.  

```python
# Feature engineered: route
# create new columne `route` by concatenate start station name and end station name
df["route"] = df['start station name']+ "_to_" + df['end station name']
```

```python
# Get route list, order in terms of its popularity, in year 2019, 2020
route_counts2019 = df.query('year == 2019')['route'].value_counts()
route_counts2020 = df.query('year == 2020')['route'].value_counts()

# get the order of route (top 10 only)
num_of_toplist = 10
route_order2019 = route_counts2019.index[:num_of_toplist]
route_order2020 = route_counts2020.index[:num_of_toplist]

# get the porportion of top most popular route
max_prop_route2019 = route_counts2019[0]/df.query('year==2019').shape[0]
max_prop_route2020 = route_counts2020[0]/df.query('year==2020').shape[0]
```

```python
# Union top 10 route list for year 2019 and 2020
route_list = route_counts2019.index[:num_of_toplist].to_list()

# loop to add in list from 2020, but not in 2019 top 20 
for route in route_counts2020.index[:num_of_toplist]:
    if route not in route_counts2019.index[:num_of_toplist]:
        route_list.append(route)
```

```python
# produce a set of evenly space proportion
tick_props = np.arange(0, max_prop_route2020+0.002, 0.002)
tick_names = ['{:0.3f}'.format(v) for v in tick_props]
```

```python
plt.figure(figsize = [15,6])
# main title
plt.subplots_adjust(top = 0.8)
plt.suptitle("Top 10 Popular Route\n(start station to end station)", fontsize = 20)

# Year 2019
plt.subplot(1,2,1)
ax1 = sns.countplot(data = df.query('year == 2019'), 
                    y = "route", palette= ['#ff9999'],
                    order = route_order2019);

ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)

# label the frequency at the end of each bar
for i in range(num_of_toplist):
    count = route_counts2019[i]
    pct_string = '{:0.1f}%'.format(100*count/df.query('year==2019').shape[0])
    plt.text(count-(count*0.4), i, pct_string, va = "center")
    
#xticks label
plt.xticks(tick_props*df.query('year==2019').shape[0], tick_names, rotation = 45)
plt.xlabel("Proportion")
plt.ylabel("")
plt.title("Year 2019", fontsize = 16);

# Year 2020
plt.subplot(1,2,2)
ax2 = sns.countplot(data = df.query('year == 2020'), 
                    y = "route", palette=['#66b3ff'],
                    order = route_order2020);

ax2.spines["top"].set_visible(False)
ax2.spines["right"].set_visible(False)

# label the frequency at the end of each bar
for i in range(num_of_toplist):
    count = route_counts2020[i]
    pct_string = '{:0.1f}%'.format(100*count/df.query('year==2020').shape[0])
    plt.text(count+1, i, pct_string, va = "center")
    
#xticks label
plt.xticks(tick_props*df.query('year==2020').shape[0], tick_names, rotation = 45)
plt.xlabel("Proportion")
plt.ylabel("")
plt.title("Year 2020", fontsize = 16);
```
