### 3.2 Assess
After gathering each of the above pieces of data, I assess them visually and programmatically for quality and tidiness issues.

#### 3.2.1 Quality issues:
This refers to data with content's problem, such as missing data, inaccurate data, inconsistent data, incorrect data type. 
- Quality issues observed are incorrect data type for example the date data in string format instead of datetime, tweet_id in integer format instead of string format.
- Inaccurate data detected for `rating_numerator` and `rating_denominator`. Some of the ratings with decimal values were incorrectly extracted. Inaccurate data also detected in `name` column, some of them are not dog's name

#### 3.2.2 Tidiness issues 
This refers to messy datas with structural problem. Tidy data should have the following characteristics: 
- Each variable forms a column
- Each observation forms a row
- Each type of observational unit forms a table.

Some examples of issues observed in the dataset:
- `dog stages` should be a single column instead of four, which violate the requirement of each variable forms a column.
- All the 3 datasets refer to same observational unit, that is everything related to tweet's post. Thus, these 3 datasets should be merged into 1 as they are part of the same observation unit (tweets).

### 3.3 Clean
Each data quality and data tidiness issues detected in assessing step were cleaned. Each item was defined, coded and tested to ensure the issue is addressed correctly.  

#### 3.3.1 Quality issues:
An example of the **data quality** issue is the incorrect data in rating related column. Some of the rating with decimal values are incorrectly extracted. The current pipeline captures incorrect values when rating numerators contact decimals. As such, I clean it by extracting the ratings again from the `text` column from the `twitter_archive` table. 
This is done by using the regex to extract the rating with either:
``` 
13/10 (without decimal)
or
9.75/10.0 (with decimal)
```
and group them by the `/` seperator into `numerator` and `denominal`.

A [useful tool](https://regex101.com/) recommended by my reviewer is a great tool for me to find the correct regex to do the above task. The regex that I used showed below:
```
ratings = twitter_clean.text.str.extract('(\d+\.?\d+|\d+)\/(\d+\.?\d+|\d+)', 
                               expand=True)
```
