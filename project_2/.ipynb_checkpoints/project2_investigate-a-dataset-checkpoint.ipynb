>

### 2.2 Data manupulation
Change the data types, add column with engineered data

```python
# change the `ScheduledDay` and `AppointmentDay` datatype to datetime datatype using 
# np.datetime64 format

df['ScheduledDay'] = df.ScheduledDay.apply(np.datetime64)
df['AppointmentDay'] = df.AppointmentDay.apply(np.datetime64)
```

```python
# get the day of the week from both `ScheduledDay` and `AppointmentDay`, 
# using .isoweekday(), where Monday = 1, Sunday = 7

df['Scheduled_weekday'] = df.ScheduledDay.dt.date.apply(lambda x:x.isoweekday())
df['Appointment_weekday'] = df.AppointmentDay.dt.date.apply(lambda x:x.isoweekday())

# find how many days appointment is made in advance of the scheduled day (waiting time)
df['days_toAppointment'] = (df.AppointmentDay - df.ScheduledDay).dt.days
```

```python
# create another column to list the day of scheduled & appointment in str
weekday_name = ['start_form_1','Mon','Tue','Wed','Thur','Fri','Sat','Sun']
df['ScheduledDay_str'] = [weekday_name[x] for x in df.Scheduled_weekday]
df['AppointmentDay_str'] = [weekday_name[x] for x in df.Appointment_weekday]
```

```python
# Change the patient id's datatype from float to string, for patient ID analysis
df['PatientId'] = df['PatientId'].apply(lambda x: str(x).split('.')[0])
```

```python
# replace category variable of 'No-show' to binary
# No-show : No -> 0
# No-show : Yes -> 1
df['No-show'] = df['No-show'].map({'No': 0, 'Yes':1})
```

```python
# drop AppointmentID as this is just a record with no additional information provided 
# in my analysis
df.drop('AppointmentID', axis=1, inplace = True)
```

It was found out that, the `ScheduledDay` and `AppointmentDay` are in the **same** day, but due to there is no time info in the `AppointmentDay`, which cause it to set it to 00:00:00 time when I find the timedelta between the 2 dates, resulted in -1 days.
```python
# For days_toAppointment = -1
# This is due to the Appointment day is only in date, no time (Hour, Minute, seconds) information, but 
# ScheduledDay has the info for both date & time. Thus, it resulted to be all in time(00:00:00) for AppointmentDay
# Since both ScheduledDay & AppointmentDay on the same day, to change days_toAppointment to 0

df[df['days_toAppointment'] == -1][['ScheduledDay', 'AppointmentDay', 'days_toAppointment']]
```

### 3.2 Is waiting time, the day of appointment, no sms reminder increase the chances of No-show?

```python
# Function for countplot
def count_pct_plot(df, x, hue_col, hue_title, title, pct_data):
    plt.subplots(figsize=(10,6))
    ax = sns.countplot(x = x, hue = hue_col, data = df, palette = "Set2")
    
    #print the percentage of no-show on each bar
    for p in ax.patches:
        x = p.get_bbox().get_points()[:,0]
        y = p.get_bbox().get_points()[1,1]
        ax.annotate('{:.2f}%'.format(100.*y/df.shape[0]), (x.mean(),y), ha= 'center', va = 'bottom')

    ax.legend(loc = 3).set_title(hue_title)
    ax.set_ylabel("Number of occurences", fontsize=12)
        
    ax2 = ax.twinx()
    ax2.plot(pct_data[:-1], color = "darkblue", marker = 'x', label = "No-show % ")
    ax2.set_ylim(0, pct_data.max()+10)
    ax2.legend(loc=1, frameon = False)
    ax2.set_ylabel("No-show percentage, %\n(# of No-show/total within same group)", fontsize=12)
    
    ax.set_title(title, fontsize=14)
```

```python
# Function to get the percentage data
def get_pct(df,x):
    
    temp = pd.crosstab(df['No-show'], df[x], margins = True, margins_name = 'Total')
    pct_per_x = temp.loc[1]/temp.loc["Total"]*100
    
    print("Percentage groupby {}".format(x))
    print(df[x].value_counts(normalize=True)*100)
    print()
    print("No-show percentage groupby:")
    print(pct_per_x)
    return pct_per_x
```

```python
# Function to plot heatmap
def plot_heatmap(data, figsize, title):
    mask = np.zeros_like(data)
    mask[np.triu_indices_from(mask)] = True
    
    with sns.axes_style("white"):
        f,ax = plt.subplots(figsize=figsize)
        ax = sns.heatmap(data, mask=mask, vmax=1, vmin=-1, annot=True,
                    linewidths = .5, cmap = "YlGnBu")
        ax.set_title(title, fontsize=14)
```

### 3.4 Any health unit (neighborhood) having higher No-show rate?

#### 3.4.1 Appointment by neighbourhood distribution
To find out the total number of appointment made by patient for each health unit and their corresponding `No-show` rate (percentage).   

The analysis and visualisation observed that:
- There are total of **81** health units for the total of **110521** scheduled appointments. However, the number of scheduled appointment varied from the heatlth unit, with as low as 1 up to 7717. 
- The average `No-show` percentage is **20.8%**. 
- **ILHA DO BOI** is with lowest `No-show` rate of **8.6%** (3/35), whereas **ILHAS OCEÂNICAS DE TRINDADE** has the highest `No-show` rate of **100%** (2/2). But, these two health units are relatively with much lower scheduled appointment.
- **JARDIM CAMBURI** is the health unit with largest number of scheduled appointment and its' `No-show` percentage is **19%**, ~ 1% point lower than the average `No-show` percentage.
- **75 percentile** of the `No-show` is from 8.6% up to **21.8%**, thus the average of 20.8% is skewed by the **"outlier"** of one health unit of having 100% `No-show` rate. 

```python
# create another dataframe to count the number of appointment made by each health unit (Neighbourhood)
nh_group = df.groupby(['Neighbourhood','No-show']).size().reset_index(name='count')
```

```python
# count the sum of appointment by each neighborhood
total = nh_group.groupby('Neighbourhood')['count'].sum()

# create another dataframe that merge the nh_group & total, to get the total column
nh_grouptotal = nh_group[nh_group['No-show'] == 1].merge(total, how='inner', on = 'Neighbourhood')
nh_grouptotal.rename(columns = {"count_x": "No-show_count", "count_y": "Total_count"}, inplace = True)
nh_grouptotal.drop('No-show', axis=1, inplace = True)

# count the percentage of No-show for each neighborhood
nh_grouptotal["No-show_pct"] = nh_grouptotal['No-show_count']/nh_grouptotal['Total_count']

# sort by neighborhood by number of patient (total_count)
nh_grouptotal.sort_values(by='Total_count',inplace=True)
nh_grouptotal.reset_index(drop=True, inplace=True)
nh_grouptotal.head()
```

```python
# Get the statistic description of the data.
# There are actually 81 unique Neighbourhood. The one missing here is 'PARQUE INDUSTRIAL', where it only has 
# 1 appointment, where the patient is from 7-18 age-group
nh_grouptotal.describe()
```

```python
# Health unit (neighborhood) with lowest No-show percentage:
nh_grouptotal.iloc[nh_grouptotal['No-show_pct'].idxmin()]
```

#### 3.4.2 Appointment by Neighbourhood and by patient's age group distribution
This section is to find out, for each health unit (Neighbourhood), what is the `No-show` rate for each age group of patient.    
The visualisation plot in later code-cell shows the health unit from the least (top in the plot) down to the max number of scheduled appointments made by patient.   
`Overall No-show` is the `No-show` percentage per health unit (that is the same data shown in the bar chart in 3.4.1). It follows by the `No-show` rate for each age group within the health unit. (e.g., there are total of 10 scheduled appointment from 0-2 age-group and 2 of them is No-show, thus the No-show rate for this age-group is 20%).

**Observation from the plot**:
- **Senior** age-group (65-115) in general has relatively **lower** `No-show` rate across **all** health units.
- Overall majority patient is from **Mid-age** group (40-64), which is found in earlier section. This age-group also has relatively lower `No-show` rate across **all** heath units, except the one health unit with 100% `No-show` rate, which all of them (2 appointments) came from this age group.
- For those health units with low scheduled appointments (those appeared top in the plot), the age-group with higher `No-show` rate is scatter among age-group <40 years old (Blue, cyan, dacker green). These age-group also has higher mean `No-show` rate. (descriptive statistic).
- For those health units with high scheduled appointments, for example the last 10 rows, the `No-show` is relatively more evenly spread (chart with mostly light green).
- Overall, there is strong positive correlation for the overall `No-show` rate to the `No-show` rate from **40-64** age group by health unit. This is due to, in general, majority patient is from this age-group and thus, has higher weights on the overall `No-show` rate. 

```python
# Get the number of appointment, break-down by Neighbourhood, age group and No-show status
nh_age = df.groupby(["Neighbourhood", "age_group", "No-show"]).size()
nh_age = nh_age.unstack(-1).unstack(-1)
nh_age
```

```python
# Create another dataframe to calculate the No-show percentage for each neighbourhood break-down by age group

nh_age_noshowpct = nh_age.loc[:,1]/(nh_age.loc[:,0]+nh_age.loc[:,1])
nh_age_noshowpct.fillna(0, inplace=True)  # fill NaN to 0
nh_age_noshowpct.head()
```

```python
# Remove the multi-index
nh_age_noshowpct.columns = nh_age_noshowpct.columns.astype('str')
nh_age_noshowpct.reset_index(inplace =True)
nh_age_noshowpct.head()
```

```python
# Merge it with nh_grouptotal dataframe that created earlier
nh_combined = nh_grouptotal.merge(nh_age_noshowpct, how="inner", on = "Neighbourhood")
nh_combined.drop(['No-show_count', 'Total_count'], axis=1, inplace=True)
nh_combined.set_index('Neighbourhood', inplace=True)
nh_combined.rename(columns = {"No-show_pct": "Overall No-show"}, inplace=True)
nh_combined.head()
```

```python
# Statistical description of No-show rate (overall and per age-group)
# 3-6, 7-18, 29-14, 25-39 in general has higher mean No-show rate (>20%)
# senior (65-115) has the lowest mean No-show rate (14.1%)
nh_combined.describe()
```

**Visualisation** of the Overall `No-show` percentage for each health unit (Neighbourhood) and the `No-show` rate from each age-group

```python
# Visualisation through heatmap
fig, ax = plt.subplots(figsize=(10, 25))
fig.subplots_adjust(top=.95)
plt.suptitle('No-show percentage by neighborhood and age group', fontsize=18)
cbar_kws = {'orientation':"horizontal", 'pad':0.08, 'aspect':50}

ax = sns.heatmap(nh_combined, annot=True, linewidths=.3, cmap='GnBu', cbar_kws=cbar_kws);
ax.set_xlabel("Age group", fontsize=12);
ax.set_ylabel("Health unit (Neighbourhood)", fontsize=12);
```

```python
# calculate the correlation matrix of No-show rate per age-group, and the Overall No-show percentage
# The overall No-show percentage has strong positive correlation with No-show rate of 40-64 age-group.
corr = nh_combined.corr()
# Visualise through heatmap
plot_heatmap(corr, (14,8), "Correlation matrix of patient's age-group and No-show percentage")
```
